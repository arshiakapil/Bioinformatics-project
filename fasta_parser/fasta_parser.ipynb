```python
def fasta_parser(fastafile):
    store_data = dict()
    with open(fastafile, "r") as f:
        for line in f.readlines():
            line = line.strip()
            if line.startswith(">"):
                getSeqID = line[1:].split()
                #print(getSeqID)
                seqID = getSeqID[0]
                #print(seqID)
                val = ''
                store_data[seqID] = ''
            else:
                val += line
                store_data[seqID] = val
    return store_data
                
                
fasta_parser("example.fasta")                
            
        
        
```
